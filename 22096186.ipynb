```python
df['Text']= df['Text'].str.strip()
df['Sentiment']= df['Sentiment'].str.strip()
df['User']= df['User'].str.strip()
df['Platform']= df['Platform'].str.strip()
df['Hashtags']= df['Hashtags'].str.strip()
df['Country']= df['Country'].str.strip()
```

```python
max_hour = df['Hour'].max()
min_hour = df['Hour'].min()
print(max_hour)
print(min_hour)
```

```python
platform_counts = df['Platform'].value_counts()
print(platform_counts)
```

```python
country_counts = df['Country'].value_counts()
print(country_counts)
```

```python
plt.figure(figsize=(20, 20))
plt.subplot(2, 2, 1)
platform_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, legend=False)
plt.title('ID#22096186\nNAME: Hashir Ibtisam\nThis plot shows the Distribution of Platforms for social media usage\nand percentage of instagram usage is highest followed Twitter and Facebook')

#plt.show()
plt.grid()

plt.subplot(2, 2, 2)
#plt.figure(figsize=(12, 6))
country_counts.plot(kind='bar', color='skyblue')
plt.title('This plot shows the countries with the percentage of social media platforms usage')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Country percentage')
plt.ylabel('Count of countries')
plt.grid()


plt.subplot(2, 2, 3)
posts_per_year = df.groupby('Year').size()

# Plot a line chart
#plt.figure(figsize=(12, 6))
posts_per_year.plot(kind='line', marker='o', color='red')
plt.title('This plot shows the Number of Posts per Year and the trend is increasing over the years')
plt.xlabel('Posts per Year')
plt.ylabel('Number of Posts over years')
plt.grid(True)


plt.subplot(2, 2, 4)
#plt.figure(figsize=(10, 6))
sns.histplot(df['Likes'], kde=True, color='blue')
plt.title('This plot shows the Distribution of Likes and it is uniform or normal')
plt.xlabel('Likes Distribution')
plt.ylabel('Frequency Distribution' )
#plt.show()
plt.grid()
plt.savefig(r'C:\Users\HP\Desktop\22096186.png',dpi=(300),bbox_inches='tight')
```
